```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
def plot_covar(sigma_xx, sigma_yy, sigma_xy):
    mean = [0, 0]
    cov = [[sigma_xx, sigma_xy], [sigma_xy, sigma_yy]]
    x, y = np.random.multivariate_normal(mean, cov, 5000).T
    fig, ax = plt.subplots(3,1, figsize=(4,6))
    plt.suptitle("Sigma_xx: %d, Sigma_yy: %d, Sigma_xy: %d" %(sigma_xx, sigma_yy, sigma_xy))
    ax[0].plot(x,y,'.')
    ax[0].axis('equal')
    ax[0].set_ylim((-12,12))
    ax[0].set_xlim((-12,12))
    ax[1].hist(x);
    ax[1].set_ylim((0,170))
    ax[1].set_xlim((-12,12))
    ax[2].set_ylim((0,170))
    ax[2].set_xlim((-12,12))
    ax[2].hist(y);
    plt.tight_layout()
    
    return fig
```

```python
from ipywidgets import StaticInteract, RangeWidget
StaticInteract(plot_covar, sigma_xx = RangeWidget(0,15,5), sigma_yy = RangeWidget(0,15,5), sigma_xy = RangeWidget(-10,20,10))    
```

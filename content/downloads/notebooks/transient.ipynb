```python
import numpy.linalg.linalg as LA
import numpy as np
import scipy as sp
from scipy.sparse.linalg import lsqr
```

```python
def create_A(n, mu):
    """
    Create the A matrix
    """
    d1=np.eye(n-1,n)
    d1=np.roll(d1,1)
    d2=-np.eye(n-1,n)
    a_second=  np.sqrt(mu)*(d1+d2)
    a_first=np.eye(n)
    A=np.vstack((a_first,a_second))
    return sp.sparse.csr_matrix(A)

def create_b(corr, n):
    b_2=np.zeros((n-1,1))
    b=np.vstack((corr,b_2))
    
    return b

def denoise(corr, mu):
    """
    Return denoised signal
    """
    n =  len(corr)
    A = create_A(n, mu)
    b = create_b(corr, n)
    return lsqr(A,b)[0]
```

```python
v1 = denoise(df_mains.head(5000).values, 1)
v2 = denoise(df_mains.head(5000).values,100)
```

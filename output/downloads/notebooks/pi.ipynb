So everyone likes to compute the value of Pi to illustrate Monte Carlo methods. I'll do the same. However, I'll be using interactive widgets to make the visualization more dynamic!

Quoting wikipedia:
1. Draw a square on the ground, then inscribe a circle within it.
2. Uniformly scatter some objects of uniform size (grains of rice or sand) over the square.
3. Count the number of objects inside the circle and the total number of objects.
4. The ratio of the two counts is an estimate of the ratio of the two areas, which is π/4. Multiply the result by 4 to estimate π.

There are some really good tutorials around the net on this topic, so I'll get straight away to the code!

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
from ipywidgets import StaticInteract, RangeWidget
import math
```

```python
def monte_carlo_pi(log_iterations):
    iterations = math.pow(10, log_iterations)
    X = np.random.uniform(size=(iterations,2))
    # Get x^2 + y^2
    s = X[:,0]*X[:,0] + X[:,1]*X[:,1]
    # Indices where x^2 + y^2 <=1 (belong inside the circle)
    idx_circle = s<=1
    # Number of points within the circle
    count = sum(idx_circle)
    fig, ax = plt.subplots(figsize=(4,3))  
    ax.set_ylim((0,1))
    ax.set_xlim((0,1))
    # Plotting the points within the circle
    ax.plot(X[idx_circle][:,0],X[idx_circle][:,1],'bo')
    # Plotting points outside circle in green
    ax.plot(X[~idx_circle][:,0],X[~idx_circle][:,1],'go')  
    # Plotting the boundaries of the circle
    x_circle = np.linspace(0,1,1000)
    y_circle = np.sqrt(1-x_circle*x_circle)
    ax.plot(x_circle, y_circle, 'r', linewidth=5)
    ax.set_title(r'Iterations: %d  $\pi$ = %0.4f' %(iterations, count*4.0/iterations))
    return fig    
```

```python
StaticInteract(monte_carlo_pi, log_iterations = RangeWidget(1,5,1))
```

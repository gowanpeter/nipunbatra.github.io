```python
import pystan
```

```python
beronoulli_model = """data {
int<lower=0> M; //Number of observation sets
int<lower=0> N; //Number of observations in each set
int<lower=0> y[M]; //Number of heads in each observation set
}
parameters {
real<lower=0,upper=1> theta_a;
real<lower=0,upper=1> theta_b;
real<lower=0, upper=1> coin_choice[M];
}
transformed parameters {
real<lower=0, upper=1> theta[M]; 
for(i in 1:M){
    if (coin_choice[i] > 0.5) 
         theta[i] <- theta_a;
    else
        theta[i] <- theta_b;
 }
}
model {
theta_a ~ beta(2,2);
theta_b ~ beta(2,2);
for(i in 1:M)
    coin_choice[i] ~ uniform(0, 1);
y ~ binomial(N,theta);
}"""
```

```python
coin_dat = {'N': 10,'M': 5, 'y': [5,9,8,7,4]}
```

```python
fit = pystan.stan(model_code=beronoulli_model, data=coin_dat, iter=500, chains=1)
```

## Lag plots

The utility of lag plots is descibed on [this](http://www.itl.nist.gov/div898/handbook/eda/section3/lagplot.htm) NIST page.

>A lag plot checks whether a data set or time series is random or not. Random data should not exhibit any identifiable structure in the lag plot. Non-random structure in the lag plot indicates that the underlying data are not random.

In these lag plots, I'll plot the value of power for different appliances/meters with lag 1. This means that if x is power at time `t`, corresponding y is power at time `t+1`. 

```python
for meter in elec.meters:
    meter.plot_lag(1)
    plt.title(meter.appliance_label())
```

Some pretty shapes, right?

## Autocorrelation plots

Again, I'll pick up an excerpt from a NIST [page](http://www.itl.nist.gov/div898/handbook/eda/section3/autocopl.htm)

>Autocorrelation plots are a commonly-used tool for checking randomness in a data set. This randomness is ascertained by computing autocorrelations for data values at varying time lags. If random, such autocorrelations should be near zero for any and all time-lag separations. If non-random, then one or more of the autocorrelations will be significantly non-zero.

```python
for meter in elec.meters:
    meter.plot_autocorrelation()
    plt.title(meter.appliance_label())
```

In the plots above, we have
>The horizontal lines displayed in the plot correspond to 95% and 99% confidence bands. The dashed line is 99% confidence band.

## Spectrum plots

>A spectral plot ( Jenkins and Watts 1968 or Bloomfield 1976) is a graphical technique for examining cyclic structure in the frequency domain. It is a smoothed Fourier transform of the autocovariance function.
The frequency is measured in cycles per unit time where unit time is defined to be the distance between 2 points. A frequency of 0 corresponds to an infinite cycle while a frequency of 0.5 corresponds to a cycle of 2 data points.

```python
for meter in elec.meters:
    meter.plot_spectrum()
    plt.title(meter.appliance_label())
```
